```python
import pandas as pd 

import matplotlib.pyplot as plt 

import seaborn as sns 
```

```python
## Import dataset
Insurance = pd.read_csv('/Users/Yedou/Desktop/SeabornInsurance/insurance.csv')
Insurance
```

```python
Insurance.head()
```

```python
# Represent a scatter plot showing the relationship between the bmi and expenses columns of our dataframe

plt.figure(figsize = (10,6))

sns.scatterplot(x = Insurance['bmi'], y = Insurance['expenses'])

plt.show()
```

```python
# Represent a scatter plot showing the relationship smoker, between, bmi and expenses columns of our dataframe

plt.figure(figsize = (10,6))

sns.scatterplot(x = Insurance['bmi'], y = Insurance['expenses'], hue = Insurance['smoker'])

plt.show()
```
